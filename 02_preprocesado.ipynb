```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train = pd.read_csv("train.csv")
print ("shape of loaded dataframe", train.shape)
```

```python
train.head()
```

Librerias


```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

Identificamos las columnas

```python
y = train["RENDIMIENTO_GLOBAL"]
X = train.drop(columns=["RENDIMIENTO_GLOBAL"])

print("Columnas:", len(X.columns))
```

```python
numeric_cols = ["INDICADOR_1", "INDICADOR_2", "INDICADOR_3", "INDICADOR_4"]
categorical_cols = [c for c in X.columns if c not in numeric_cols]

print("Categóricas:", len(categorical_cols))
print("Numéricas:", len(numeric_cols))
```

Se rellenan los datos faltantes

```python
X[numeric_cols] = X[numeric_cols].fillna(X[numeric_cols].median())
```

```python
X[categorical_cols] = X[categorical_cols].fillna("MISSING")
```

```python
X.head()
```

Convertimos variables categoricas a numeros

```python
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le

X.head()
```
